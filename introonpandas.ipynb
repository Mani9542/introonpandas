```python
# prompt: import matploty liberary to write commands

import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('salary_data.csv')
#df #This will print the entire dataframe.  Consider removing or limiting the output.
#df.head() #This will print the first 5 rows. Consider removing or limiting the output
#df.tail() #This will print the last 5 rows. Consider removing or limiting the output

#df.shape #Prints the shape of the dataframe


df=pd.read_csv('salary_data.csv')
#null_values = df.isnull().sum()
#null_values #Prints the number of null values in each column

#df.isnull() #Prints a boolean dataframe showing null values
#df.describe() #Prints summary statistics for the dataframe

# Example of a matplotlib plot
# Assuming your salary_data.csv has a 'YearsExperience' and a 'Salary' column
plt.figure(figsize=(8, 6))
plt.scatter(df['YearsExperience'], df['Salary'])  # Create a scatter plot
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary vs. Years of Experience")
plt.grid(True)
plt.show()
```
